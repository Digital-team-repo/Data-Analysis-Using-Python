## Objective
Import an Excel file and examine the distribution of school enrolment values.

## Dataset
- Source page: https://data.ontario.ca/dataset/school-information-and-student-demographics
- Classroom Excel file: `school-demographics-classroom.xlsx`


## Step 1: Import libraries
Big picture: We load the tools needed to open Excel data and draw a chart.

Technical details: `pandas` creates a DataFrame named `df`. `plotly.express` builds the chart.

```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the Excel file
Big picture: We open the classroom-ready Excel file stored in this same folder.

Technical details: `pd.read_excel()` reads the spreadsheet into `df`.

```python
data_url = "https://github.com/Digital-team-repo/Data-Analysis-Using-Python/raw/refs/heads/main/04-excel-analysis/school-demographics-classroom.xlsx"
df = pd.read_excel(data_url)
df.head()
```

## Step 3: Check your table
Big picture: We confirm the columns before plotting.

Technical details: `df.columns` is a property, while `df.head()` is a method.

```python
print(df.columns)
print(df.shape)
```

## Step 4: Filter to secondary schools (Process step)
Big picture: Before plotting, we clean and focus the table so our chart answers a clear question about secondary schools only.

Technical details: We use boolean filtering with `df[df['School Level'] == 'Secondary']` and store the result in `secondary_df`.

```python
secondary_df = df[df['School Level'] == 'Secondary'].copy()
print(secondary_df.shape)
secondary_df.head()
```

## Step 5: Create the chart
Big picture: We create one clear visualization to spot a pattern quickly.

Technical details: The histogram uses `Enrolment` values from `secondary_df` to show how common different secondary-school sizes are.

```python
fig = px.histogram(secondary_df, x='Enrolment', nbins=30, title='Distribution of Secondary School Enrolment')
fig.show()
```
