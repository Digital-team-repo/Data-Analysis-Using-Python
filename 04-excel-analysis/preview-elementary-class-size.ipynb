## Objective
Import an Excel file and compare class-size distributions across grades.

## Dataset
- Source page: https://data.ontario.ca/dataset/elementary-class-size
- Classroom Excel file: `elementary-class-size-classroom.xlsx`


## Step 1: Import libraries
Big picture: We load the tools needed to open Excel data and draw a chart.

Technical details: `pandas` creates a DataFrame named `df`. `plotly.express` builds the chart.

```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the Excel file
Big picture: We open the classroom-ready Excel file stored in this same folder.

Technical details: `pd.read_excel()` reads the spreadsheet into `df`.

```python
data_file = "https://github.com/Digital-team-repo/Data-Analysis-Using-Python/raw/refs/heads/main/04-excel-analysis/elementary-class-size-classroom.xlsx"
df = pd.read_excel(data_file)
df.head()
```

## Step 3: Check your table
Big picture: We confirm the columns before plotting.

Technical details: `df.columns` is a property, while `df.head()` is a method.

```python
df.columns
df.shape
```

## Step 4: Create the chart
Big picture: We create one clear visualization to spot a pattern quickly.

Technical details: The box plot uses `Grade` on x-axis and `Class Size` on y-axis after reshaping the table.

```python
long_df = df.melt(id_vars=['SCHOOLYEAR', 'BOARDNAME', 'SCHOOLNAME', 'CLASS_NUMBER'], value_vars=['JK', 'K', 'G1', 'G2', 'G3', 'G4TO8'], var_name='Grade', value_name='Class Size')
long_df = long_df.dropna(subset=['Class Size'])
fig = px.box(long_df, x='Grade', y='Class Size', title='Elementary Class Size Distribution by Grade')
fig.update_layout(template='plotly_white')
fig.show()
```
