## Objective
Import an Excel file and view weekly price trends over time.

## Dataset
- Source page: https://data.ontario.ca/dataset/average-weekly-cattle-prices
- Classroom Excel file: `average-weekly-cattle-prices-classroom.xlsx`


## Step 1: Import libraries
Big picture: We load the tools needed to open Excel data and draw a chart.

Technical details: `pandas` creates a DataFrame named `df`. `plotly.express` builds the chart.

```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the Excel file
Big picture: We open the classroom-ready Excel file stored in this same folder.

Technical details: `pd.read_excel()` reads the spreadsheet into `df`.

```python
data_url = "https://github.com/Digital-team-repo/Data-Analysis-Using-Python/raw/refs/heads/main/04-excel-analysis/average-weekly-cattle-prices-classroom.xlsx"
df = pd.read_excel(data_url)
df.head()
```

## Step 3: Check your table
Big picture: We confirm the columns before plotting.

Technical details: `df.columns` is a property, while `df.head()` is a method.

```python
print(df.columns)
print(df.shape)
```

## Step 4: Create the chart
Big picture: We create one clear visualization to spot a pattern quickly.

Technical details: The line chart uses `Week_Ending` on x-axis and two price columns on y-axis. `900-999 lbs` and `over 1250 lbs` are cattle weight ranges in pounds, and prices are in dollars per cwt (live weight).

```python
fig = px.line(df, x='Week_Ending', y=['Steer_900_999_lbs_Price_per_cwt_live', 'Steer_over_1250_lbs_Price_per_cwt_live'], title='Average Weekly Cattle Prices Over Time')
fig.for_each_trace(lambda t: t.update(name='Steer 900-999 lbs') if t.name == 'Steer_900_999_lbs_Price_per_cwt_live' else t.update(name='Steer over 1250 lbs'))
fig.update_layout(
    yaxis_title='Price ($ per cwt, live weight)',
    legend_title_text='Cattle Group',
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='left', x=0, font=dict(size=13))
)
fig.show()
```
