## Objective
Import an Excel file, inspect more rows, and compare histogram distributions before and after filtering.

## Dataset
- Source page: https://data.ontario.ca/dataset/school-information-and-student-demographics
- Classroom Excel file: `data_en.xlsx`


## Step 1: Import libraries
Big picture: We load the tools needed to open Excel data and draw a chart.

Technical details: `pandas` creates a DataFrame named `df`. `plotly.express` builds the chart.

```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the Excel file
Big picture: We open the classroom-ready Excel file stored in this same folder.

Technical details: `pd.read_excel()` reads the spreadsheet into `df`.

```python
data_url = "https://github.com/Digital-team-repo/Data-Analysis-Using-Python/raw/refs/heads/main/04-excel-analysis/data_en.xlsx"
df = pd.read_excel(data_url)
df.head()
```

## Step 3: Check your table
Big picture: We inspect structure and row count before making charts.

Technical details: `df.columns` is a property, `df.head(10)` shows the first 10 rows, `df.tail(10)` shows the last 10 rows, and `len(df)` returns row count.

```python
print(df.columns)
```

```python
print(df.shape)
```

```python
print(len(df))
```

```python
print(df.head(10))
```

```python
print(df.tail(10))
```

## Step 4: Create a histogram with all schools
Big picture: We first chart the full dataset so we have a baseline view of enrolment distribution.

Technical details: A histogram groups enrolment values into bins and shows how many schools fall in each range.

```python
fig_all = px.histogram(df, x='Enrolment', nbins=30, title='Distribution of School Enrolment (All Schools)')
fig_all.show()
```

## Step 5: Filter to secondary schools (Process step)
Big picture: Now we narrow the dataset to answer a more specific question.

Technical details: We use boolean filtering with `df[df['School Level'] == 'Secondary']` and store results in `secondary_df`.

```python
secondary_df = df[df['School Level'] == 'Secondary'].copy()
print(secondary_df.shape)
print(len(secondary_df))
secondary_df.head(10)
secondary_df.tail(10)
```

## Step 6: Create a histogram with secondary schools only
Big picture: We compare this second chart with the all-schools chart to see how filtering changes the distribution.

Technical details: This histogram uses the same `Enrolment` column, but only from `secondary_df`.

```python
fig_secondary = px.histogram(secondary_df, x='Enrolment', nbins=30, title='Distribution of School Enrolment (Secondary Schools Only)')
fig_secondary.show()
```
