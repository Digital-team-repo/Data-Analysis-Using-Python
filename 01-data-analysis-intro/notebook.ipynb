**Objective:** Use a simple temperature example to understand the IPO workflow (Input, Process, Output) in a notebook.

**Dataset description:** This file shows annual average temperature departure values for Canada (difference from the 1961-1990 normal) from 1948 to 2024.

**Data source:** This version uses the **cleaned CSV file** (`data_en.csv`) located in this folder, adapted from the original dataset on the [Open Government Portal](https://open.canada.ca/data/en/dataset/8b01db78-b825-4d38-a66c-1a1712c382ee).

## Step 1 (Process Setup): Import libraries
This setup step runs first so we can perform the Input step next (`pd.read_csv(...)`), and later complete the Process work.

```python
import pandas as pd
import plotly.express as px
```

## Step 2 (Input): Load the data file
We load `data_en.csv` from this folder as the Input for our workflow.

```python
df = pd.read_csv("data_en.csv")
df.head()
```

## Step 3 (Process): Inspect columns
We check variable names before plotting so the Process step is intentional.

```python
df.columns
```

## Step 4 (Output): Create a visualization
This chart is our Output: a visual summary of the data trend over time.

```python
fig = px.scatter(
    df, 
    x="Year", 
    y="Temperature departure (degree Celsius)",
    title="Annual Temperature Change in Canada (1948-2024)",
    trendline="ols"
)

fig.show()
```
