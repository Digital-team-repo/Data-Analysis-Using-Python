# Lesson 1: Temperature Change Exploration

**Objective:** Explore trends in Canada's annual average temperature departures using a cleaned dataset and scatter plots.

**Dataset description:** This file shows the annual average temperature departure (difference from the 1961-1990 normal) for Canada from 1948 to 2024.

**Data source(s):**
- Original Dataset: https://open.canada.ca/data/en/dataset/8b01db78-b825-4d38-a66c-1a1712c382ee
- This version uses a **cleaned CSV file** (`data_en.csv`) where the government header rows have been removed for easier analysis.

**Date/author:** 2026-05-14, Antigravity (AI Assistant)


## Step 1: Import our libraries

In Python, `import` loads tools from libraries so we can use them in our notebook.

- `pandas` (as `pd`) helps us load and work with table data.
- `plotly.express` (as `px`) helps us create charts quickly.


```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the CSV and perform a "First Look"

We load the file into a DataFrame called `df` and check the first few rows.


```python
url = "data_en.csv"
df = pd.read_csv(url)
df.head()
```

## Step 3: Display the column names


```python
df.columns
```

## Step 4: Build a scatter plot

We will create a scatter plot to show how temperature departures have changed over time.


```python
fig = px.scatter(df, x="Year", y="Temperature departure (degree Celsius)", 
                 title="Annual Temperature Departures in Canada (1948-2024)",
                 trendline="ols")
fig.show()
```
