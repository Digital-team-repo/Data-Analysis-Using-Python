**Objective:** Explore trends in Canada's annual average temperature departures using the provided cleaned dataset and scatter plots.

**Dataset description:** This file shows the annual average temperature departure (difference from the 1961-1990 normal) for Canada from 1948 to 2024.

**Data source:** This version uses the **cleaned CSV file** (`data_en.csv`) located in this folder, adapted from the original dataset on the [Government of Canada Temperature Change indicator page](https://www.canada.ca/en/environment-climate-change/services/environmental-indicators/temperature-change.html).

## Step 1: Import libraries
Import `pandas` for data handling and `plotly.express` for visualization.

```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the cleaned data file
We load `data_en.csv` directly from this folder.

```python
data_url = "https://raw.githubusercontent.com/Digital-team-repo/Data-Analysis-Using-Python/refs/heads/main/03-csv-analysis/data_en.csv"

df = pd.read_csv(data_url)
df.head()
```

### Quick Check: `df.head()` vs `df.tail()`
`df.head()` shows the first 5 rows of the table so we can quickly check whether the file loaded correctly and whether the columns look right.

## Step 4: Create a Scatter Plot
Visualize the temperature departure over time with a trendline.

```python
fig = px.scatter(
    df, 
    x="Year", 
    y="Temperature departure (degree Celsius)",
    title="Annual Temperature Change in Canada (1948-2024)",
    trendline="ols"
)

fig.show()
```
