# Lesson 1: Extreme Heat Events Exploration

**Objective:** Explore trends in extreme heat events in Canada using a cleaned dataset and beginner-friendly charts.

**Dataset description:** This file summarizes long-term Canadian extreme heat trends (1948 to 2023). Each row represents a trend category, with counts (`Number of Stations`) and percentages (`Percentage`) describing how many monitoring stations fit that trend.

**Data source(s):**
- Original Dataset: https://open.canada.ca/data/en/dataset/2cc1ad59-97cf-4652-839a-7cfaa943ebcf
- This version uses a **cleaned CSV file** (`data_en.csv`) where the government header rows have been removed for easier analysis.

**Date/author:** 2026-05-14, Antigravity (AI Assistant)


## Step 1: Import our libraries

In Python, `import` loads tools from libraries so we can use them in our notebook.

- `pandas` (as `pd`) helps us load and work with table data.
- `plotly.express` (as `px`) helps us create charts quickly.

We run this setup cell once near the top of the notebook.


```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the CSV and perform a "First Look"

Professional analysts never jump straight to making charts. First, they need to "shake hands" with the data to understand its shape and labels. We call this a **First Look**.

This cell does three things in order:
1. Stores the dataset URL in a variable called `url`.
2. Uses `pd.read_csv(url)` to load the file into a DataFrame called `df`.
3. Shows the first few rows with `df.head()`. You can also use `df.tail()` to see the very end of the file.

**Quick vocabulary:**
- A **DataFrame** is pandas' table object (rows and columns, like a spreadsheet).
- We name it `df` as a short common name for "data frame."
- In `df.head()` or `df.tail()`, the part after the dot is a **method** (an action).


```python
url = "https://raw.githubusercontent.com/Digital-team-repo/Data-Analysis-Using-Python/refs/heads/lesson-setup/01-extreme-heat/data_en.csv"
df = pd.read_csv(url)
df.head()
```

## Step 3: Display the column names

Before graphing, we first list the column names so we know exactly what fields are available in the dataset. 

In our IPO model, this is like checking the labels on our ingredients before we start cooking. If we don't know the column names, we can't tell the computer which data to process or visualize.


```python
display(df.columns)
```

## Step 4: Build a bar chart

Now we are ready to visualize the data! We will create a bar chart using `Trend` for the categories and `Number of Stations` for the values.

Notice how we use `px.bar()` to tell the computer we want a bar chart.


```python
fig = px.bar(
    df, 
    x="Trend", 
    y="Number of Stations", 
    title="Canadian Extreme Heat Trends (1948-2023)"
)

fig.show()
```
